# Generating charuco markers

```python
import cv2
import numpy as np

# 1. Define ChArUco board dimensions and parameters
squares_x = 2             # Number of squares horizontally
squares_y = 2             # Number of squares vertically
square_length = 0.04       # Length of each square (e.g., in meters)
marker_length = 0.03       # Length of the ArUco marker inside the square
dictionary_type = cv2.aruco.DICT_4X4_50  # Choose dictionary
dictionary = cv2.aruco.getPredefinedDictionary(cv2.aruco.DICT_4X4_50)
marker_size = 150
corner_labels = [
    (0, "ID=0  G3-FAR"),
    (1, "ID=1  G5-FAR"),
    (2, "ID=2  G3-NEAR"),
    (3, "ID=3  G5-NEAR"),
]

markers = []
for marker_id, label,  in corner_labels:
    marker_img = cv2.aruco.generateImageMarker(dictionary, marker_id, marker_size)
    # markers.append(cv2.cvtColor(marker_img, cv2.COLOR_GRAY2BGR))
    marker_color = cv2.cvtColor(marker_img, cv2.COLOR_GRAY2BGR)
    marker_color = cv2.copyMakeBorder(marker_color, 20, 60, 20, 20,
                                      cv2.BORDER_CONSTANT, value=(255, 255, 255))
    h, w = marker_color.shape[:2]
    cv2.putText(marker_color, label, (10, h - 15),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 0), 2)
    markers.append(marker_color)

row_top    = np.hstack([markers[0], markers[1]])   # far edge
row_bottom = np.hstack([markers[2], markers[3]])   # near edge
sheet      = np.vstack([row_top, row_bottom])

# 4. Save and display
cv2.imwrite("corner_markers.png", sheet)
cv2.imshow("Corner Markers - print and cut into 4", sheet)
cv2.imwrite("corner_markers.png", sheet)
print("Saved!")
cv2.waitKey(330)
cv2.destroyAllWindows()
```
